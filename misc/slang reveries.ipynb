```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi


def plot_voronoi_diagram(points):
  """Plots a Voronoi diagram for the given set of points.

  Args:
    points: A NumPy array of shape (N, 2) containing the coordinates of the points
      in the Voronoi diagram.
  """

  # Create the Voronoi diagram
  voronoi = Voronoi(points)

  # Create a figure
  fig, ax = plt.subplots()

  # Plot the points
  ax.scatter(points[:, 0], points[:, 1], c='black')

  # Plot the Voronoi regions
  for region in voronoi.regions:
    # Check if the region is infinite
    if -1 not in region:
      # Get the vertices of the region
      vertices = voronoi.vertices[region]

      # Plot the edges of the region
      ax.plot(vertices[:, 0], vertices[:, 1], c='blue')

  # Set the limits of the plot
  ax.set_xlim([min(points[:, 0]) - 1, max(points[:, 0]) + 1])
  ax.set_ylim([min(points[:, 1]) - 1, max(points[:, 1]) + 1])

  # Set the labels for the axes
  ax.set_xlabel('X')
  ax.set_ylabel('Y')

  # Set the title of the plot
  ax.set_title('Voronoi diagram')

  # Show the plot
  plt.show()

import numpy as np
import scipy.spatial as spatial

# Create a set of points
points = np.array([[1, 2], [3, 4], [5, 6]])

# Add a fourth point
points = np.append(points, [[0, 0]], axis=0)

# Create the Voronoi diagram
voronoi = spatial.Voronoi(points)

# Plot the Voronoi diagram
plot_voronoi_diagram(points)
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Create an array of angles
theta = np.linspace(0, 2*np.pi, 100)

# Ellipse parameters for the first ellipse
a1, b1 = 4, 2
x1 = a1 * np.cos(theta)
y1 = b1 * np.sin(theta)

# Ellipse parameters for the second ellipse
a2, b2 = 2, 4
x2 = a2 * np.cos(theta)
y2 = b2 * np.sin(theta)

# Plot the first ellipse
plt.plot(x1, y1, label='Ellipse 1')

# Plot the second ellipse
plt.plot(x2, y2, label='Ellipse 2')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Ellipses')
plt.legend()

# Show the plot
plt.show()
```
